#Mean()
Mean sendiri digunakan untuk data yang memiliki sedikit sifat outlier/noisy/anomali dalam sebaran datanya maupun isinya.

```python
# we can fill the null value use mean

data_mising.mean()
```

```python
data_fill_mean = data_mising.fillna(data_mising.mean())
print("Data yang kosong sudah terisi oleh nilai mean")
data_fill_mean.head(10)
```

#median()
median digunakan untuk data-data yang memiliki sifat outlier yang kuat. 

Kenapa median dipilih? Median merupakan nilai tengah yang artinya bukan hasil dari perhitungan yang melibatkan data outlier

```python
data_mising.median()
```

```python
data_fill_median = data_mising.fillna(data_mising.median())
print("Data yang kosong disii dengan nilai median")
data_fill_median.head(10)
```

#Normalisasi Data
image.png

kolom Usia dan Masa kerja memiliki range Puluhan sedangkan Gaji memiliki kolom range Jutaan. maka harus dinormalisasikan lebih dulu agar saat dilakukan clustering tidak menjadi masalah.

Kenapa menjadi masalah? karena klastering atau klasifikasi menggunakan k-means menggunakan proses jarak dalam menghitungnya.

Jika tidak ada normalisasi, maka jelas perhitungan kmeans diatas akan tergantung pada Gaji. Kenapa? Karena gaji berdomain jutaan dan 2 kolom lainnya hanya berdomain puluhan. Berapapun usia dan masa kerja seseorang tidak akan berpengaruh terhadap penilaian suatu perusahaan. Perbedaan skala pada setiap kolom ini merupakan hal yang sangat wajar dan sering terjadi dan inilah pentingnya normalisasi. Normalisasi sangat penting, terutama untuk yang menggunakan perhitungan jarak dengan menggunakan metode apapun.

#Metode normalisasi

##1. MinMax, 
minmax form.png
 
 Termasuk metode normalisasi yang bersifat linier dengan data aslinya. Namun, metode ini **bisa menyebabkan out of bound** pada beberapa kasus.

 Out of Bound terjadi apabila ada data baru masuk, dimana data tersebut melebihi nilai maksimal atau nilai minimal dari data yang sudah ada. Secara otomatis, perhitungan yang berlaku pada data yang sudah diperoleh tadi harus diulangi lagi semuanya dengan data baru yang masuk atau data baru yang mempunyai nilai maksimal/minimum yang melebihi tadi tidak bisa diproses. 
 
 ***tidak cocok untuk analisa real time / evolving system***
 
 ***sangat dianjurkan untuk kasus-kasus berbasis time frame analisis dan forecasting***

metode normalisasi pengembangan transformasi secara linier. Output range-nya adalah 0-1. Metode ini sangat berguna pada saat data yang ada melibatkan data outlier.

```python
from sklearn import preprocessing
array = csv_data.values

X = array[:,2:5] #memisahkan fitur dari dataset. 
Y = array[:,0:1]  #memisahkan class dari dataset

dataset=pd.DataFrame({"Customer ID":array[:,0],"Gender":array[:,1],"Age":array[:,2],"Income":array[:,3],"Spending Score":array[:,4]})
print("dataset sebelum dinormalisasi :")
print(dataset.head(10))

min_max_scaler = preprocessing.MinMaxScaler(feature_range=(0,1)) #inisialisasi normalisasi MinMax
data = min_max_scaler.fit_transform(X) #transformasi MinMax untuk fitur
dataset = pd.DataFrame({"Age":data[:,0],"Income":data[:,1],"Spending Score":data[:,2],"Customer ID":array[:,0],"Gender":array[:,1]})

print("dataset setelah dinormalisasi :")
print(dataset.head(10))
```
